```python
#Importing panda
import pandas as pd
df = pd.read_csv('D:\\Data Analysis\\Dataset\\cafe_sales_proj1.csv')
```

```python
#Making a backup copy
df_original = df.copy()
```

```python
#Preview of the data
df.head(100)
```

```python
#removing duplicates
df = df.drop_duplicates()
df
```

```python
#replacing invalid values with NA
check_here_for_invalid_values = ['Item', 'Quantity', 'Price Per Unit', 'Total Spent', 'Payment Method', 'Location', 'Transaction Date']
df[check_here_for_invalid_values] = df[check_here_for_invalid_values].replace(['UNKNOWN', 'ERROR'], pd.NA)
df.head(100)
```

```python
#Checking prices
df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'], errors='coerce')
result = df.groupby('Item')['Price Per Unit'].min().sort_values()
print(result)
```

```python
#populating the missing values or NA 'Price Per Unit' based on the price_item
price_item = {'Cookie':1.0, 'Tea':1.5,'Coffee': 2.0, 'Cake':3.0, 'Juice':3.0, 'Sandwich':4.0, 'Smoothie':4.0, 'Salad': 5.0}
df['Price Per Unit'] = df['Item'].map(price_item)
df.head(100)
```

```python
#Populating Total Spent and Quantity
#Converting to numeric first
df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'], errors='coerce')
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['Total Spent'] = pd.to_numeric(df['Total Spent'], errors='coerce')
#Computing total spent
df.loc[df['Total Spent'].isna(), 'Total Spent'] = (
    df['Price Per Unit'] * df['Quantity']
)
#Computing Quantity
df.loc[df['Quantity'].isna() & df['Price Per Unit'].notna(), 'Quantity'] = (
    df['Total Spent'] / df['Price Per Unit']
)
df.head(1000)
```

```python
# Convert columns to their proper Pandas nullable dtypes
df = df.astype({
    "Item": "string",
    "Quantity": "float",
    "Price Per Unit": "float",
    "Total Spent": "float",
    "Payment Method": "string",
    "Location": "string",
    "Transaction Date": "datetime64[ns]"
})
df
```
